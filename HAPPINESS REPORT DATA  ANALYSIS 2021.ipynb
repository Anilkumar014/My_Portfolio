```python
data_columns=['Country name', 'Regional indicator', 'Happiness Score',
              'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
              'Freedom to make life choices', 'Generosity',
              'Perceptions of corruption']
```

```python
df1=df[data_columns].copy()
```

```python
df1.columns
```

```python
#change columns name
df1.rename(columns = {'Country name':'country_name', 'Regional indicator':'regional_indicator', 'Happiness Score':'happiness_score',
              'Logged GDP per capita':'logged_gdp_per_capita', 'Social support':'social_support', 'Healthy life expectancy':'healthy_life_expectancy',
              'Freedom to make life choices':'freedom_to_make_life_choices', 'Generosity':'generosity',
              'Perceptions of corruption':'perceptions_of_corruption'}, inplace = 'True')
```

```python
#create plot between gdp happieness

plt.rcParams['figure.figsize']=(15,7)
plt.title('PLOT B/W HAPPINESS SCORE AND GDP')
sns.scatterplot(x=df1.happiness_score, y= df1.logged_gdp_per_capita, hue= df1.regional_indicator, s= 200);
plt.legend(loc='upper left', fontsize='10')

plt.xlabel('HAPPINESS SCORE')
plt.ylabel('GDP PER CAPITA')
```

```python
#GDP region wise
gdp_reg= df1.groupby('regional_indicator')['logged_gdp_per_capita'].sum()
```

```python
gdp_reg
```

```python
gdp_reg.plot.pie(autopct='%1.1f%%')
plt.title('GDP BT REGION')
plt.ylabel('')
```
